```python
x_train,x_test,y_train,y_test=train_test_split(arr,y,test_size=0.25, random_state=10)
```

```python
lm=LinearRegression()
lm.fit(x_train,y_train)
```

```python
# lets create a function to see adj-R square
def adj_R(x,y):
  r2=lm.score(x,y)
  n=x.shape[0]
  p=x.shape[1]
  adjusted_r2=1-(1-r2)*(n-1)/(n-p-1)
  return adjusted_r2
```

```python
adj_R(x_test,y_test)
```
